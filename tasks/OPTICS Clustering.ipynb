```python
from sklearn.cluster import OPTICS
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
X, y = make_blobs(n_samples=2000, centers=4, cluster_std=0.60, random_state=0)
```

```python
import matplotlib.pyplot as plt
plt.scatter( X[: , 0]  , X[: , 1] , c=y , cmap='turbo' , edgecolor='k', s=50)
plt.title("Scatter Plot of make_moons Dataset")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
optics = OPTICS(min_samples=50, xi=.05)
optics.fit(X)
```

```python
labels = optics.labels_
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='turbo')
plt.show()
```
